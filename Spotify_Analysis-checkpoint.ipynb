```python
#importing all essential libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv(r"Spotify.csv")
data
```

```python
#view columns of dataset
data.columns
```

```python
#view info of dataset
data.info()
```

```python
#dropping unessecary columns from dataset
data.drop(columns=['Unnamed: 0','Url_spotify','Uri','Url_youtube'],inplace=True)
data.info()
```

```python
#viewing no. of null values in columns
data.isna().sum()
```

```python
#filling missing/null values with 0 in Likes & Comments column
data['Likes'] = data['Likes'].fillna(0)
data['Comments'] = data['Comments'].fillna(0)
data.isna().sum()
```

```python
#to drop the null values from dataset
data.dropna(inplace = True)
data.isna().sum()
```

Q.1 Top 10 Artist with the Highest Views on Youtube

```python
Artist_views = data.groupby('Artist')['Views'].sum().sort_values(ascending = False).head(10)
Artist_views
```

Q.2 Top 10 Tracks with the Highest Stream on Spotify

```python
Track_stream = data.groupby('Track')['Stream'].sum().sort_values(ascending = False).head(10)
Track_stream
```

Q.3 What are the most common Album types on Spotify? How many Tracks belong to each Album type?

```python
data.Album_type.unique() #unique values
```

```python
A_type = data.Album_type.value_counts() #unique value with their count
A_type
```

```python
#pie chart
plt.pie(A_type, labels= A_type.index,autopct="%1.1f%%",startangle=60,colors='myg',shadow=True,explode=(0.05,0.05,0.05),pctdistance=0.75)
plt.show()
plt.savefig('Tracks.png')
```

Q.4 How do Average Likes, Views and Comments are compared between different Album Types?

```python
Avg_LVC = data.groupby('Album_type')[['Likes','Views','Comments']].mean()
Avg_LVC
```

```python
type(Avg_LVC)
```

```python
df = Avg_LVC.reset_index() #to convert the index of a series into a column to form a dataframe
df
```

```python
#melt dataframe - unpivot data
data_melted = pd.melt(df, id_vars = 'Album_type', var_name = 'Attribute', value_name = 'Total')
data_melted
```

```python
#draw barplot
plt.figure(figsize = (9,6))
sns.barplot(x='Album_type', y='Total',hue='Attribute',data=data_melted)
plt.savefig('Album.png')
```
